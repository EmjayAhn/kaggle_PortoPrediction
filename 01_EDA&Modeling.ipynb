### Column 정리

```python
data = []
for column in train_set.columns:
    if column == 'id':
        role = 'id'
    elif column == 'target':
        role = 'target'
    else:
        role = 'input'
    
    if ('bin' in column) or (column == 'target'):
        level = 'binary'
    elif ('cat' in column) or (column == 'id'):
        level = 'nominal'
    elif train_set[column].dtype == float:
        level = 'interval'
    elif train_set[column].dtype == int:
        level = 'ordinal'
        
    keep = True
    if column == 'id':
        keep = False
    
    dtype = train_set[column].dtype
    
    column_info = {
        'varname': column,
        'role': role,
        'level': level,
        'keep': keep,
        'dtype': dtype
    }
    data.append(column_info)
meta = pd.DataFrame(data, columns=['varname', 
                                   'role', 
                                   'level', 
                                   'keep', 
                                   'dtype'])
meta.set_index('varname', inplace=True)
meta
```

```python
meta[(meta['level'] == 'nominal') & (meta['keep']==True)]
```

```python
pd.DataFrame({
    'count': meta.groupby(['role', 'level'])
    ['role'].size()
}).reset_index()
```

#### Binary 변수들
- binary 변수에는 결측치가 따로 없는 것으로 보입니다.
- binary 에도 모두 unique value 가 0, 1 로 구성되어 있습니다.
- mean 값을 보게 되면 대부분이 imbalanced data 로 보입니다.

```python
v = meta[(meta['level'] == 'binary') & (meta['keep'])].index
for column in v:
    unique_value = sorted(train_set[column].unique())
    if unique_value != [0, 1]:
        print("binary 라 했는데, binary 가 아니잖아")
print("모두 binary real")

train_set[v].describe()
```

#### 결측치 확인하기

```python
missing_vars = []
for column in train_set.columns:
    missings = train_set[train_set[column] == -1][column].count()
    if missings > 0:
        missing_vars.append(column)
        missing_percent = missings / train_set.shape[0]
        print('{} : {} 개의 missing Value - {:.2%}'.format(column, missings, missing_percent))
        
print('{} Variables with missing values'.format(len(missing_vars)))
```

- `ps_car_03_cat`, `ps_car_05_cat`의 경우, 결측치가 매우 많으므로 버립니다.
- `ps_reg_03` 의 경우, continuous 하므로, 평균으로 채웁니다.
- `ps_car_11` 의 경우, ordinal 하므로, mode 로 채웁니다.
- `ps_car_12`, `ps_car_14`의 경우, continuous 하므로, 평균으로 채웁니다.
